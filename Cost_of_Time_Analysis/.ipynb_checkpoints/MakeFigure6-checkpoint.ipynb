### Make Figure 6 
Qualitative predictions of optimal policy for different cost-structures

Load Packages

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['svg.fonttype'] = 'none'
```

Define different cost functions and corresponding optimal policies

```python
# compare 3 different cost functions
plt.rc('font', size=8)

def linear_cost(a, delta_T):
    return a*delta_T

def linear_tmax(a, r):
    return (1/a)*r

def nonlinear_cost(a,b,delta_T):
    return a*np.power(delta_T,b)

def nonlinear_tmax(a,b,r):
    return np.power((1/a)*r,1/b)

def nonlinear_cost_w_horizon(a,b,T,delta_T):
    return a*np.power((1 - (1/T)*(T - delta_T)),b)

def nonlinear_cost_w_horizon(a,b,T,delta_T):
    return a*np.power(1 - (T - delta_T)/T, b)

def nonlinear_cost_w_horizon(a,b,T,delta_T):
    return a*np.power(delta_T/T, b)

def nonlinear_tmax_w_horizon(a,b,T,r):
    return T*np.power(r/a,1/b)
```

Plot these

```python
fig = plt.figure(layout='constrained', figsize=(4, 3.5), dpi=300)
subfigs = fig.subfigures(2, 1)


axsTop = subfigs[0].subplots(1, 3, sharex=True)
axsBottom = subfigs[1].subplots(1, 3, sharex=True)


delta_T = np.arange(1,30)
r = np.arange(0,.4,.005)

b = 5
T = 100
a = .05
axsTop[0].plot(delta_T, linear_cost(a,delta_T), 'k-')
axsTop[0].set_title('Linear', fontsize= 8)
axsBottom[0].plot(r, linear_tmax(a,r), 'k-')
axsBottom[0].set_title('Linear', fontsize= 8)


a = .00001
b = 4
axsTop[1].plot(delta_T, nonlinear_cost(a,b,delta_T), 'k-')
axsTop[1].set_title('Nonlinear', fontsize = 8)
axsBottom[1].plot(r, nonlinear_tmax(a,b,r), 'k-')
axsBottom[1].set_title('Nonlinear', fontsize = 8)

a = 1
b = 2
for T in [40, 50, 60]:
    axsTop[2].plot(delta_T, nonlinear_cost_w_horizon(a,b,T,delta_T), label = T)
    axsBottom[2].plot(r, nonlinear_tmax_w_horizon(a,b,T,r), label = T)

axsTop[2].legend(title = 'Time left (s)', fontsize=6, title_fontsize=6)
axsTop[2].set_title('Nonlinear with \n time left dependence', fontsize = 8)
axsBottom[2].legend(title = 'Time Left (s)', fontsize=6, title_fontsize=6)
axsBottom[2].set_title('Nonlinear with \n time left dependence', fontsize = 8)
#axsBottom[2].set_aspect('equal', 'box')

subfigs[0].supylabel('Cost of Time Spent, $c$', fontsize = 8)
subfigs[0].supxlabel('Time Spent (s), $\Delta T_C$', fontsize = 8)
# subfigs[0].suptitle('Cost Structure', fontsize=10)

#################
subfigs[1].supylabel('Max Time to Wait, $\Delta T_{max}$', fontsize = 8)
subfigs[1].supxlabel('Benefit of Computation, $\Delta U_{C}$', fontsize = 8)
```
